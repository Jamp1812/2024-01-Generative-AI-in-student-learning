```python
reasons_against_AI_use = "prefer to work myself, too bad for my subject/case, encountered serious errors, doesn’t help me, use forbidden by uni, makes up sources, ethical/moral concerns, wouldn’t learn myself, too bad for my subject/case, too unreliable, prefer to think for myself, creates bias, prefer to think for myself, prefer to think for myself, I don't trust AI, wouldn’t learn myself, I don't trust AI, prefer to work myself, wouldn’t learn myself, stealing/plagiarism, ethical/moral concerns, makes up sources, doesn’t help me, error concerns, wouldn’t learn myself, makes forget how to think for yourself, I don’t understand AI, My work can only be done by me, prefer to work myself, I don't trust AI, nobody understands exactly how AI works, error concerns, still too bad for various complex tasks, I don't trust AI, wouldn’t learn myself, don’t know how to use AI, still too bad for various complex tasks, AI does not know the truth, too unreliable, rather for generic tasks, still too bad for various complex tasks, error concerns, ok in certain cases, wouldn’t learn myself, I don't need it, error concerns, too bad for my subject/case, would feel like cheating, too unreliable, error concerns, ok in certain cases, too unreliable, use forbidden by uni, wouldn’t learn myself, Feels like doing nothing, prefer to work myself, too unreliable, error concerns, ok in certain cases, rather for generic tasks"

reasons_against_AI_use_list = reasons_against_AI_use.split(", ")
reasons_frequencies = Counter(reasons_against_AI_use_list)

# Remove entries that appear only once. They couldn't be read in the resulting plot (font size)
reasons_frequencies_processed = {key:val for key, val in reasons_frequencies.items() if val > 1}
# print(reasons_frequencies_processed)

wordcloud = WordCloud(background_color="white", relative_scaling=0.5, scale = 5).generate_from_frequencies(reasons_frequencies_processed)

plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()

# wordcloud.to_file("reasons_against_AI.pdf")
```
